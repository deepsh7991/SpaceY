First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

We can plot out the <code>FlightNumber</code> vs. <code>PayloadMass</code>and overlay the outcome of the launch. We see that as the flight number increases, the first stage is more likely to land successfully. The payload mass is also important; it seems the more massive the payload, the less likely the first stage will return.


```python
import pandas as pd

# Assuming features_df is your DataFrame containing the specified columns
features_df = pd.DataFrame({
    'FlightNumber': [1, 2, 3],
    'PayloadMass': [1000, 2000, 3000],
    'Orbit': ['LEO', 'GTO', 'ISS'],
    'LaunchSite': ['CCAFS', 'VAFB', 'KSC'],
    'Flights': [1, 2, 3],
    'GridFins': [True, False, True],
    'Reused': [False, True, False],
    'Legs': [4, 2, 3],
    'LandingPad': ['PAC', 'OCISLY', 'LZ-1'],
    'Block': [1, 2, 3],
    'ReusedCount': [0, 1, 2],
    'Serial': ['B0001', 'B0002', 'B0003']
})

# List of columns to be one-hot encoded
columns_to_encode = ['Orbit', 'LaunchSite', 'LandingPad', 'Serial']

# One-hot encode the specified columns
one_hot_encoded = pd.get_dummies(features_df[columns_to_encode])

# Calculate the total number of columns
total_columns = features_df.shape[1] + one_hot_encoded.shape[1]

print("Total number of columns after one-hot encoding:", total_columns)
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
### TASK  6: Visualize the launch success yearly trend
df['Date'] = pd.to_datetime(df['Date'])

# Extract year from the 'Date' column
df['Year'] = df['Date'].dt.year

# Group by year and calculate the success rate
yearly_success_rate = df.groupby('Year')['Class'].apply(lambda x: (x == 1).mean())

# Plot the launch success yearly trend
plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_success_rate)
plt.title('Launch Success Yearly Trend')
plt.xlabel('Year')
plt.ylabel('Success Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
def Extract_year(df):
    years = []
    for date in df['Date']:
        year = date.year  # Extract year from Timestamp object
        years.append(year)
    df['Date'] = years

# Call the function to extract years from the 'Date' column
Extract_year(df)
# Display the DataFrame with the updated 'Date' column
print(df.head())

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the 'Date' and 'Class' columns
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract year from 'Date' column
df['Year'] = df['Date'].dt.year

# Calculate success rate by year
success_rate_by_year = df.groupby('Year')['Class'].mean().reset_index()

# Plot line chart
plt.figure(figsize=(10, 6))
sns.lineplot(data=success_rate_by_year, x='Year', y='Class')
plt.title('Success Rate by Year')
plt.xlabel('Year')
plt.ylabel('Success Rate')
plt.xticks(range(2013, 2021))  # Limit x-axis to show data only from 2013 to 2020
plt.tight_layout()
plt.show()
```

you can observe that the sucess rate since 2013 kept increasing till 2020
